```python
class CSVDataset(Dataset):
    
    def __init__(self,path,transform=True):
        self.path = path
        self.df =self.__load_csv(path)
        self.transform = transform
    
    def __len__(self):
        return len(self.df)
    
    def __getitem__(self, index):
        row =  self.df.iloc[index,]   
        if self.transform:
            row['comment_text'] = [torch.randint(256,[1]) for i in row['comment_text']]
            
        return row.to_dict()
        
        
    def __load_csv(self,path):
        return pd.read_csv(path)
        
            
```

```python
dataset = CSVDataset("data/raw/sample.csv")
```

```python
BATCH_SIZE = 1
data_loader = DataLoader(dataset,batch_size = BATCH_SIZE, pin_memory=True)
for row in data_loader:
    print(row)
    
    
    
```
